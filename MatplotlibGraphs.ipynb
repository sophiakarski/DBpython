```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
has_closed = [3.5,
              3.7,
              5.1,
              7.7,
              8.0,
              9.6,
              13.8,
              22.7,
              27.0,
              29.1,
              81.2,
              82.2,
              24.3
             ]
has_continued = [95.6,
                 96.2,
                 94.4,
                 92.3,
                 92.0,
                 89.7,
                 85.8,
                 77.2,
                 72.8,
                 70.9,
                 18.4,
                 16.7,
                 75.4
                ]
```

```
industry_labels = ['Science And Tech',
                   'Education',
                    'ICT',
                    'Manufacturing',
                    'Water And Waste Management',
                    'Admin And Support Service',
                    'Wholesale/Retail Trade; Vehicle Repair',
                    'Arts Entertainment And Recreation',
                    'Construction',
                    'Transportation And Storage',
                    'Health And Social Work',
                    'Accommodation And Food',
                    'All Industries'
                   ]
```

```
industry_label_position = np.arange(len(industry_labels))
```

```
#setting bars
width = 0.35
fig , ax = plt.subplots()
closed_bars = ax.bar(industry_label_position - width / 2,
                     has_closed,
                     width,
                     label = 'closed')

continued_bars = ax.bar(industry_label_position + width / 2,
                     has_continued,
                     width,
                     label = 'continued')

#setting labels
ax.set_ylabel('percentage')
ax.set_xticks(industry_label_position)
ax.set_xticklabels(industry_labels)
ax.set_title('% of businesses closing or continuing')
ax.legend()

#rotate industry label
plt.xticks(rotation = 90)


plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt


N = 13
Lower = (95.2, 96.3, 93.4, 84.9, 95.1, 94.5, 81.1, 97.2, 96.5, 94.4, 98.1, 92.1, 90.2)
Higher = (1.7, 0.6, 3.6, 13.2, 0.0, 3.4, 18.0, 2.9, 1.4, 5.0, 1.9, 7.9, 8.0)
Unsure = (3.0, 3.1, 2.9, 1.9, 4.9, 2.1, 1.0, 0.0, 2.1, 0.5, 0.0, 0.0, 1.8)
ind = np.arange(N)    # the x locations for the groups
width = 0.65       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, Lower, width)
p2 = plt.bar(ind, Higher, width,
             bottom=Lower)
p3 = plt.bar(ind, Unsure, width)
             


plt.ylabel('Percentage')
plt.title('Businesses reporting turnover rates compared to normal')
plt.xticks(ind, ('Science And Tech',
                   'Education',
                    'ICT',
                    'Manufacturing',
                    'Water And Waste Management',
                    'Admin And Support Service',
                    'Wholesale/Retail Trade; Vehicle Repair',
                    'Arts Entertainment And Recreation',
                    'Construction',
                    'Transportation And Storage',
                    'Health And Social Work',
                    'Accommodation And Food',
                    'All Industries'))

#rotate industry label
plt.xticks(rotation = 90)

plt.yticks(np.arange(0, 110, 10))
plt.legend((p1[0], p2[0], p3[0]), ('Lower', 'Higher', 'Unsure'))

plt.show()
```
