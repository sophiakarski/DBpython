```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
has_closed = [3.5,
              3.7,
              5.1,
              7.7,
              8.0,
              9.6,
              13.8,
              22.7,
              27.0,
              29.1,
              81.2,
              82.2,
              24.3
             ]
has_continued = [95.6,
                 96.2,
                 94.4,
                 92.3,
                 92.0,
                 89.7,
                 85.8,
                 77.2,
                 72.8,
                 70.9,
                 18.4,
                 16.7,
                 75.4
                ]
```

```
industry_labels = ['Science And Tech',
                    'Education',
                    'ICT',
                    'Manufacturing',
                    'Water And Waste Management',
                    'Admin And Support Service',
                    'Wholesale/Retail Trade; Vehicle Repair',
                    'Arts Entertainment And Recreation',
                    'Construction',
                    'Transportation And Storage',
                    'Health And Social Work',
                    'Accommodation And Food',
                    'All Industries'
                   ]
```

```
industry_label_position = np.arange(len(industry_labels))
```

```
#setting bars
width = 0.35
fig , ax = plt.subplots()
closed_bars = ax.bar(industry_label_position - width / 2,
                     has_closed,
                     width,
                     label = 'closed')

continued_bars = ax.bar(industry_label_position + width / 2,
                     has_continued,
                     width,
                     label = 'continued')

#setting labels
ax.set_ylabel('percentage')
ax.set_xticks(industry_label_position)
ax.set_xticklabels(industry_labels)
ax.set_title('% of businesses closing or continuing')
ax.legend()

#rotate industry label
plt.xticks(rotation = 90)


plt.show()
```
